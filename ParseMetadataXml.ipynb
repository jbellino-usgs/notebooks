```python
from lxml import etree

root = etree.parse('ds926_fig17_top_UpperDolostoneUnit_pts.xml').getroot()
print(etree.tostring(root, pretty_print=True))
```

###Define some data dictionaries to use when rewriting xml tag text

```python
main_citation = {'pubdate': '20150304',
                 'title': 'DS926 Digital surfaces and thicknesses of selected hydrogeologic units of the Floridan aquifer system in Florida and parts of Georgia, Alabama, and South Carolina -- Points for the top of the Avon Park dolostone',
                 'geoform': 'vector digital data',
                 'onlink': 'http://water.usgs.gov/lookup/getspatial?ds926_fig17_top_UpperDolostoneUnit_pts'
                }
larger_work_citation = {'pubdate': '20150304',
                        'title': 'Digital Surfaces and Thicknesses of Selected Hydrogeologic Units of the Floridan Aquifer System in Florida and Parts of Georgia, Alabama, and South Carolina',
                        'geoform': 'Publication',
                        'onlink': 'http://pubs.usgs.gov/ds/0926/'
                       }
descript = {'purpose': }
```

###Iterate over the XML document and replace tags as defined in the dictionaries above

```python
stream = []
for element in root.iter():
    if isinstance(element.tag, basestring):
        stream.append(element.tag)
        
        # Edit title, geoform, and onlink of the main citation
        if 'lworkcit' not in stream:
            
            if element.tag in main_citation.keys():
                element.text = main_citation[element.tag]
        
        # Edit title, geoform, and onlink of the main larger work citation
        else:
            
            if element.tag in larger_work_citation.keys():
                element.text = main_citation[element.tag]
                
        
        # Remove the extra "S" character at the beginning of this section
        if element.tag == 'purpose':
            element.text = element.text[1:]
                
#             print element.getparent().tag
#         print("%s - %s" % (element.tag, element.text))
```

###Write the new XML file

```python
etree.ElementTree(root).write('test.xml')
with open('test.xml', 'r') as f:
    lines = f.readlines()
    
with open('test.xml', 'w') as f:
    f.write('<?xml version="1.0" encoding="UTF-8"?>\n')
    for line in lines:
        f.write(line)
```
