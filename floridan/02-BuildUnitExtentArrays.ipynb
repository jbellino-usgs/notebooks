#Build Unit-Extent Arrays
This notebook is used to define the extent of each of the units to be used to establish layering in the Floridan model. The output is a series of numpy arrays, of shape (nrow, ncol), that define these extents.

###Import modules and setup workspace

```python
import os
import arcpy
import matplotlib.pyplot as plt
import paths
import fasgas_plotter
import fasgas_utils

# Set some arcpy variables
arcpy.CheckOutExtension("Spatial")
arcpy.env.overwriteOutput = True

# This is where the framework surfaces are stored
arcpy.env.workspace = r'..\gis\unit_extents\extents.gdb'

# This is the model grid to be used
grid = r'..\gis\grids\grid_5000ft_v3\grid.shp'

# Read metadata associated with the model grid
ul_x, ul_y, delx, dely, nrow, ncol, rotation, proj4 = fasgas_utils.import_grid_metadata(grid)

# Specify the output director
outdir = r'dat\extent'

# Format of the output ascii grid values
fmt = '%d'

# Make a feature layer upon which selections can be made
grid_lyr = r'in_memory\grid_lyr'
arcpy.MakeFeatureLayer_management(grid, grid_lyr)
```

###Make extent for entire Floridan aquifer system

```python
# Select grid cells whose centroid falls within boundary features
arcpy.SelectLayerByLocation_management(grid_lyr, 
                                       overlap_type='HAVE_THEIR_CENTER_IN', 
                                       select_features='fas_extent', 
                                       selection_type='NEW_SELECTION')

# Create a blank array
ext_FAS = np.zeros((nrow, ncol), dtype=np.int)

# Set active
ext_FAS = fasgas_utils.set_val(grid_lyr, ext_FAS, val=1)

np.savetxt(os.path.join(outdir,'FASGAS_EXT_FAS.dat'), ext_FAS, fmt='%d')

fasgas_plotter.plot_image(ext_FAS, 'Extent of FAS', os.path.join(outdir,r'png\FASGAS_EXT_FAS.png'))
```

###Make unit extent array for surficial aquifer system


```python
# Select grid cells whose centroid falls within boundary features, set inactive where FAS outcrops
arcpy.SelectLayerByLocation_management(grid_lyr, 
                                       overlap_type='HAVE_THEIR_CENTER_IN', 
                                       select_features='fas_extent', 
                                       selection_type='NEW_SELECTION')

arcpy.SelectLayerByLocation_management(grid_lyr, 
                                       overlap_type='HAVE_THEIR_CENTER_IN', 
                                       select_features='fas_outcrops', 
                                       selection_type='REMOVE_FROM_SELECTION')

# Create a blank array
ext_SAS = np.zeros((nrow, ncol), dtype=np.int)

# Set active
ext_SAS = fasgas_utils.set_val(grid_lyr, ext_SAS, val=1)

np.savetxt(os.path.join(outdir,'FASGAS_EXT_SAS.dat'), ext_SAS, fmt='%d')

fasgas_plotter.plot_image(ext_SAS, 'Extent of SAS', os.path.join(outdir, r'png\FASGAS_EXT_SAS.png'))
```
